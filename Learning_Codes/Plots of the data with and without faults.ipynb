### Visualisation of faultless data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
file_name = "waterTank_Golden_raw.csv"
# file_name = "test_series_original.csv"
attr = list(pd.read_csv(file_name).columns.values)[1:]
# df_raw = pd.read_csv(file_name, index_col=0, header=0, parse_dates=True)
df_raw = pd.read_csv(file_name, index_col=0, header=0, parse_dates=True)
df_raw.index.freq = 'ms'
df_raw = df_raw.values[0:2000]

# plt.title('3 Water Tank Sensor Data')
ylabel = 'Sensor values'
xlabel = 'Timesteps'
num_variables = 10

# Create a 5x2 subplot grid for 10 variables
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(10,12))

#Flatten the axes array to easily loop over each subplot
axes = axes.flatten()

# Loop through each variable and plot in the corresponding subplot
for i in range(num_variables):
    axes[i].plot(df_raw[:, i])
#     axes[i].set_title(f'Variable {i + 1}')
    axes[i].set_title(attr[i])
    axes[i].set_xlabel(xlabel)
    axes[i].set_ylabel(ylabel)

# Remove any unused subplots (if the number of variables is less than 10)
for i in range(num_variables, len(axes)):
    fig.delaxes(axes[i])
    
# Adjust spacing between subplots
plt.tight_layout()

plt.show()
```
